### Determining the mean and SD to be used for generating the random data
- HEalth data has expected reference ranges and the mean and SD of these ranges will be utikized in generating random data

```python
def calculate_mean_std(low_limit, high_limit):
    mean = (low_limit + high_limit) / 2
    std_dev = (mean - low_limit) / 2
    return mean, std_dev
```

```python
low = 60
high =100

mean, std_dev = calculate_mean_std(low, high)

print(f"Mean: {mean}, Standard Deviation: {std_dev}")
```

### Define average resting heart based on age and hpertension == 1

1. Infants (1 month-1 year): 80-160 bpm

2. Childhood (1-12 years): 75 - 130 bpm

3. Others ≥ 13 years): 60-100 bpm

4. Where hypertension == 1: > 80 bpm


```python
# Simulating resting heart rate (RHR) based on normal refernece age RHR mean and SD and hypertension status
def simulate_RHR(df):
    if df['age'] <= 1:
        RHR = np.random.normal(120, 20)  # mean and SD1 year and below
    elif df['age'] > 1 and df['age'] <= 12:
      RHR = np.random.normal(90, 20)  # mean and SD for > 1 to 12 years
    else:
       RHR = np.random.normal(80, 10)   # > 12 years and above
    
    # Adjust for hypertension
    if df['hypertension'] == 1:
       RHR += np.random.uniform(5, 15) # higher RHR for patients with hypertension

   # Set realistic limits for RHR
       RHR = max(60, min(RHR, 200))
    
    return round(RHR, 1)
```

```python
# Apply the function to your DataFrame
healthdata['resting_heart_rate'] = healthdata.apply(simulate_RHR, axis=1)
```
